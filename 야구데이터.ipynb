```python
import requests
import time
# url = 'http://www.kbreport.com/leader/main' #실제 url
url = 'http://www.kbreport.com/leader/list/ajax' #ajax 요청
param = {"rows":"20",
         "order":"oWAR",
         "orderType":"DESC",
         "teamId":"",
         "defense_no":"",
         "year_from":"2020",
         "year_to":"2020",
         "gameType":"",
         "split01":"",
         "split02_1":"",
         "split02_2":"",
         "r_tpa_count":"",
         "tpa_count":"",
         "page":1}

pd_list = []
page = 1
while True:
    print(page, end=', ')
    param['page']=page
    res = requests.post(url, data=param)
    if res.status_code == 200:
        df = pd.read_html(res.text, index_col='#')[0]
        if df.shape[0] == 0: #조회결과가 없으면 행이 (0, 20) shape이다.
            break
        pd_list.append(df)
        page+=1
        time.sleep(1)
        
```
