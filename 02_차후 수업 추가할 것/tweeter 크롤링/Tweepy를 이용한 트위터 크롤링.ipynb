```python
import pickle
with open('keys.pkl', 'rb') as f:
    keys = pickle.load(f)  
```

```python
# ì¸ì¦ì„ ìœ„í•œ cunsumer_key ë“±ë¡
CONSUMER_KEY = keys['cunsumer_key']
CONSUMER_SECRET = keys['cunsumer_secret_key']

# accessë¥¼ ìœ„í•œ access_token ë“±ë¡
ACCESS_TOKEN = keys['access_token']
ACCESS_TOKEN_SECRET =keys['access_token_secret']
```

```python
# ì¸ì¦ ë°›ê¸°
auth = OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
# access í† í° ìš”ì²­
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

# API ê°ì²´ ìƒì„±. 
api = API(auth)
```

```python
# í‚¤ì›Œë“œ ì…ë ¥ ë°›ì•„ íŠ¸ìœ—ê²€ìƒ‰
keyword = input("í‚¤ì›Œë“œ:")
```

```python
max_tweets = 5
searched_tweets = []
last_id = -1
while len(searched_tweets) < max_tweets:
    count = max_tweets - len(searched_tweets)
    print("count:%d" % count)
    try:
        new_tweets = api.search(q=keyword, count=count) #??íŒŒë¼ë¯¸í„° count ì—­í• ?
        
        if not new_tweets:
            break
        searched_tweets.extend(new_tweets)
#         last_id = new_tweets.id
        
    except tweepy.TweepError as e:
        print(e)
        break
        
```
