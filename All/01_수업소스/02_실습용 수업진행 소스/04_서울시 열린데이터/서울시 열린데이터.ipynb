## 서울시 공공자전거 실시간 대여정보
- https://data.seoul.go.kr/dataList/OA-15493/A/1/datasetView.do
- 공공데이터 메뉴 클릭해서 나옴

```python
api_key = None
with open('api_key.txt') as f:
    api_key = f.read()
print(api_key)   
```

```python
import requests
from pprint import pprint
json_url = "http://openapi.seoul.go.kr:8088/{key}/json/bikeList/{start}/{end}/"
url = json_url.format(key=api_key, start=1, end=200)
print(url)
res = requests.get(url)
if res.status_code == 200:
    data = res.json()
    print(type(data))
    pprint(data)
```

```python
rows = data['rentBikeStatus']['row']
for row in rows:
    print(row['stationName'])
```
