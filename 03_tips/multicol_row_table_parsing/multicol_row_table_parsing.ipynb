```python
from bs4 import BeautifulSoup

bs = BeautifulSoup(open('test.html'), 'lxml')

bs.prettify()
```

```python
table = bs.select_one('div#layerRow_1 > table')
table
```

# 행의 multi index 조회

```python
depth_3 = None
depth_2 = None
depth_1 = None

td_list = bs.select('div#layerRow_1 > table > tbody > tr > td')
print(len(td_list))
row_index_result = []
for td in td_list:
    coldepth = td.get('coldepth') # 대/중/소 분류 구분 : 3 - 대, 2 - 중, 1 - 소
    if coldepth == '3':
        depth_3 = td.get_text()
    elif coldepth == '2':
        depth_2 = td.get_text()
    elif coldepth == '1':
        depth_1 = td.get_text()
        # row_index_result.append(','.join([depth_3, depth_2, depth_1]))
        row_index_result.append([depth_3, depth_2, depth_1])
```

```python
from pprint import pprint
pprint(row_index_result)
```

# 데이터 조회

```python
data_tr_list = bs.select('div#layerData_1 > table > tbody > tr')
data_result = []
for tr in data_tr_list:
    td_list = tr.select('td')
    # tmp = ','.join(td.get_text() for td in td_list)
    data_result.append([td.get_text() for td in td_list])
```

```python
print(data_result)
```

```python
len(row_index_result), len(data_result)
```

```python
import csv
with open("result.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(['대분류', '중분류', '소분류', '데이터', 'col1', 'col2', 'col3', 'col4', 'col5', 'col6', 'col7', 'col8', 'col9', 'col10', 'col11', 'col12', 'col13', 'col14', 'col15', 'col16', 'col17', 'col18', 'col19'])

    for row_index, data in zip(row_index_result, data_result):
        # print(row_index+","+data)
        writer.writerow(row_index + data)
```
